```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import re
import seaborn as sns
import emojis
from collections import Counter
```

```python
# –ü–µ—Ä–≤—ã–π —á–∞—Ç

f = open('result_close.json', 'r', encoding='utf-8')
data_1 = json.load(f)
f.close()

# –í—ã–≤–µ–¥–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ
print(data_1['name'])

# –°–ª–æ–≤–∞—Ä—å –¥–ª—è –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
d_close = {'id':[],'from':[], 'type':[], 'text':[], 'date':[]}

for message in data_1['messages']:

    if message['type'] == 'message':
        d_close['id'].append(message['id'])
        d_close['type'].append(message['type'])
        d_close['date'].append(message['date'])
        d_close['from'].append(message['from'])
        if 'text' in message:
            d_close['text'].append(message['text'])

        else:
            d_close['text'].append('')
    else:
      d_close['id'].append(message['id'])
      d_close['type'].append(message['type'])
      d_close['date'].append(message['date'])
      d_close['from'].append(message['actor'])
      d_close['text'].append('')

# –°–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –∏–∑ —Å–ª–æ–≤–∞—Ä—è
df_close = pd.DataFrame(d_close)
df_close
```

```python
# –ß–µ—Ç–≤—ë—Ä—Ç—ã–π —á–∞—Ç

f = open('result_school.json', 'r', encoding='utf-8')
data_4 = json.load(f)
f.close()

# –í—ã–≤–µ–¥–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ
print(data_4['name'])

# –°–ª–æ–≤–∞—Ä—å –¥–ª—è –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
d_school = {'id':[],'from':[], 'type':[], 'text':[], 'date':[]}

for message in data_4['messages']:

    if message['type'] == 'message':
        d_school['id'].append(message['id'])
        d_school['type'].append(message['type'])
        d_school['date'].append(message['date'])
        d_school['from'].append(message['from'])
        if 'text' in message:
            d_school['text'].append(message['text'])

        else:
            d_school['text'].append('')
    else:
      d_school['id'].append(message['id'])
      d_school['type'].append(message['type'])
      d_school['date'].append(message['date'])
      d_school['from'].append(message['actor'])
      d_school['text'].append('')

# –°–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –∏–∑ —Å–ª–æ–≤–∞—Ä—è
df_school = pd.DataFrame(d_school)

df_school
```

```python
# –ü—è—Ç—ã–π —á–∞—Ç

f = open('result_speech.json', 'r', encoding='utf-8')
data_5 = json.load(f)
f.close()

# –í—ã–≤–µ–¥–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ
print(data_5['name'])

# –°–ª–æ–≤–∞—Ä—å –¥–ª—è –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
d_speech = {'id':[],'from':[], 'type':[], 'text':[], 'date':[]}

for message in data_5['messages']:

    if message['type'] == 'message':
        d_speech['id'].append(message['id'])
        d_speech['type'].append(message['type'])
        d_speech['date'].append(message['date'])
        d_speech['from'].append(message['from'])
        if 'text' in message:
            d_speech['text'].append(message['text'])

        else:
            d_speech['text'].append('')
    else:
      d_speech['id'].append(message['id'])
      d_speech['type'].append(message['type'])
      d_speech['date'].append(message['date'])
      d_speech['from'].append(message['actor'])
      d_speech['text'].append('')

# –°–æ–∑–¥–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –∏–∑ —Å–ª–æ–≤–∞—Ä—è
df_speech = pd.DataFrame(d_speech)
df_speech
```

```python
# –û–±—ä–µ–¥–∏–Ω—è–µ–º –ø—è—Ç—å —á–∞—Ç–æ–≤ –≤ –æ–¥–∏–Ω –¥–∞—Ç–∞—Å–µ—Ç

df_otec = pd.concat([df_close, df_club, df_school, df_krutotec, df_speech], ignore_index=True)
df_otec
```

–û—á–∏—Å—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–æ–≤—ã—Ö –∫–æ–ª–æ–Ω–æ–∫

```python
# –£–¥–∞–ª–∏–º —Å–µ—Ä–≤–∏—Å–Ω—ã–µ —Å–æ–æ–±—â–µ–Ω–∏—è
df_otec = df_otec[df_otec["type"] == "message"]

# –£–¥–∞–ª–∏–º —Å—Ç—Ä–æ–∫–∏ –±–µ–∑ —Ç–µ–∫—Å—Ç–∞
df_otec = df_otec.dropna(subset=["text"]).copy()

# –£–¥–∞–ª–∏–º —Å—Å—ã–ª–∫–∏ –∏–∑ df –∏ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º list –≤ str
for idx in df_otec.index:
    if type(df_otec.loc[idx, 'text']) == list:
        new_string = ''
        for elem in df_otec.loc[idx, 'text']:
            if type(elem) == str:
                new_string += elem
        df_otec.loc[idx, 'text'] = new_string

# –î–æ–±–∞–≤–∏–º –Ω–æ–≤—ã–µ –∫–æ–ª–æ–Ω–∫–∏ –ø–æ –¥–∞—Ç–µ
df_otec['date'] = pd.to_datetime(df_otec['date'], format='%Y-%m-%dT%H:%M:%S')
df_otec['year'] = df_otec['date'].dt.year
df_otec['month'] = df_otec['date'].dt.month
df_otec['day'] = df_otec['date'].dt.day_name()
df_otec['hour'] = df_otec['date'].dt.hour

# –î–æ–±–∞–≤–∏–º –∫–æ–ª–æ–Ω–∫—É –ø–æ –¥–ª–∏–Ω–µ —Å–æ–æ–±—â–µ–Ω–∏—è
df_otec["message_length"] = df_otec["text"].astype(str).apply(len)

# –°—á–∏—Ç–∞–µ–º —ç–º–æ–¥–∑–∏ –≤ –æ—Ç–¥–µ–ª—å–Ω—É—é –∫–æ–ª–æ–Ω–∫—É, –ø–æ–¥—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ
df_otec['count_emoji'] = df_otec['text'].apply(lambda s: emojis.count(str(s)))
df_otec['emoji'] = df_otec['text'].apply(lambda s: [em for em in emojis.iter(str(s))])

df_otec
```

–î–∞—Ç–∞—Å–µ—Ç –≥–æ—Ç–æ–≤, —Å–æ—Ö—Ä–∞–Ω–∏–º –µ–≥–æ –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π —Ñ–∞–π–ª –Ω–∞ –≤—Å—è–∫–∏–π —Å–ª—É—á–∞–π

# –ê–Ω–∞–ª–∏—Ç–∏–∫–∞

–°–¥–µ–ª–∞–µ–º –ø–µ—Ä–≤–∏—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑

```python
# 1. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–æ–æ–±—â–µ–Ω–∏–π –ø–æ –º–µ—Å—è—Ü–∞–º, –¥–Ω—è–º –∏ —á–∞—Å–∞–º
messages_per_month = df_otec.groupby("month").size()
messages_per_day = df_otec.groupby("day").size()
messages_per_hour = df_otec.groupby("hour").size()

# –°—Ä–µ–¥–Ω–µ–µ –∏ –º–µ–¥–∏–∞–Ω–∞ —Å–æ–æ–±—â–µ–Ω–∏–π –≤ –º–µ—Å—è—Ü, –¥–µ–Ω—å, —á–∞—Å
mean_messages_per_month = messages_per_month.mean()
median_messages_per_month = messages_per_month.median()
mean_messages_per_day = messages_per_day.mean()
median_messages_per_day = messages_per_day.median()
mean_messages_per_hour = messages_per_hour.mean()
median_messages_per_hour = messages_per_hour.median()

# 2. –°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–º–æ–¥–∑–∏ –≤ —Å–æ–æ–±—â–µ–Ω–∏–∏
mean_emoji_per_message = df_otec["count_emoji"].mean()

# 3. –ê–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
user_message_counts = df_otec["from"].value_counts()
mean_messages_per_user = user_message_counts.mean()
median_messages_per_user = user_message_counts.median()

# 4. –î–ª–∏–Ω–∞ —Å–æ–æ–±—â–µ–Ω–∏–π
mean_message_length = df_otec["message_length"].mean()
median_message_length = df_otec["message_length"].median()
```

```python
# –ê–Ω–∞–ª–∏–∑ —ç–º–æ–¥–∑–∏
# 1. –§–∏–ª—å—Ç—Ä—É–µ–º —Ç–æ–ª—å–∫–æ —Å—Ç—Ä–æ–∫–∏, –≥–¥–µ –µ—Å—Ç—å —ç–º–æ–¥–∑–∏
df_emojis = df_otec[df_otec["count_emoji"] > 0].copy()

# 2. –°–æ–±–∏—Ä–∞–µ–º –≤—Å–µ —ç–º–æ–¥–∑–∏ –≤ –æ–¥–∏–Ω —Å–ø–∏—Å–æ–∫
all_emojis = [emoji for sublist in df_emojis["emoji"] for emoji in sublist]

# 3. –ü–æ–¥—Å—á—ë—Ç —á–∞—Å—Ç–æ—Ç—ã —ç–º–æ–¥–∑–∏
emoji_counts = Counter(all_emojis)
top_10_emojis = emoji_counts.most_common(10)
```

```python
# === –í—ã–≤–æ–¥ –æ—Å–Ω–æ–≤–Ω—ã—Ö –º–µ—Ç—Ä–∏–∫ ===
print("–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–æ–±—â–µ–Ω–∏–π –≤ –º–µ—Å—è—Ü:", mean_messages_per_month)
print("–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–æ–±—â–µ–Ω–∏–π –≤ –º–µ—Å—è—Ü:", median_messages_per_month)
print("–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–æ–±—â–µ–Ω–∏–π –≤ –¥–µ–Ω—å:", mean_messages_per_day)
print("–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–æ–±—â–µ–Ω–∏–π –≤ –¥–µ–Ω—å:", median_messages_per_day)
print("–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–æ–±—â–µ–Ω–∏–π –≤ —á–∞—Å:", mean_messages_per_hour)
print("–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–æ–±—â–µ–Ω–∏–π –≤ —á–∞—Å:", median_messages_per_hour)
print("–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–º–æ–¥–∑–∏ –≤ —Å–æ–æ–±—â–µ–Ω–∏–∏:", mean_emoji_per_message)
print("–°—Ä–µ–¥–Ω–µ–µ —Å–æ–æ–±—â–µ–Ω–∏–π –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è:", mean_messages_per_user)
print("–ú–µ–¥–∏–∞–Ω–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–π –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è:", median_messages_per_user)
print("–°—Ä–µ–¥–Ω—è—è –¥–ª–∏–Ω–∞ —Å–æ–æ–±—â–µ–Ω–∏–π:", mean_message_length)
print("–ú–µ–¥–∏–∞–Ω–Ω–∞—è –¥–ª–∏–Ω–∞ —Å–æ–æ–±—â–µ–Ω–∏–π:", median_message_length)
```

–°—Ä–µ–¥–∏ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è —Å–∞–º–∏ –≤–ª–∞–¥–µ–ª—å—Ü—ã —á–∞—Ç–æ–≤ –∏ –∫–∞–Ω–∞–ª—ã. –ü–æ—á–∏—Å—Ç–∏–º –∏ –≤—ã–≤–µ–¥–µ–º –Ω–æ–≤—É—é –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—é

```python
# –°–ø–∏—Å–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã—Ö –Ω—É–∂–Ω–æ –∏—Å–∫–ª—é—á–∏—Ç—å
excluded_users = [
    "–û–¢–¶–´ –†–Ø–î–û–ú online",
    "–û—Ç—Ü–æ–≤—Å–∫–∏–π –∫–ª—É–± –®–∫–æ–ª—ã –º—É–∂–µ—Å—Ç–≤–∞",
    "–ö—Ä—É—Ç–û—Ç–µ—Ü –ß–∞—Ç",
    "–ö—Ä—É—Ç–û—Ç–µ—Ü",
    "–ß–∞—Ç - –û—Ç—Ü–æ–≤—Å–∫–∏–π —Ä–∞–∑–≥–æ–≤–æ—Ä",
    "–û—Ç—Ü–æ–≤—Å–∫–∏–π –∫–ª—É–± –®–∫–æ–ª—ã –º—É–∂–µ—Å—Ç–≤–∞ Chat",
    "–û—Ç–µ—Ü —Å–≤–æ–µ–π –ø—Å–∏—Ö–æ–ª–æ–≥–∏–∏ | –ò–ª—å—è –õ–æ–±–æ–¥–∏–Ω"
]

# –§–∏–ª—å—Ç—Ä—É–µ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
filtered_user_message_counts = user_message_counts[~user_message_counts.index.isin(excluded_users)]

# –ü–µ—Ä–µ—Å—á–∏—Ç—ã–≤–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∏ –º–µ–¥–∏–∞–Ω—É –±–µ–∑ –∏—Å–∫–ª—é—á–µ–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
mean_messages_per_user = filtered_user_message_counts.mean()
median_messages_per_user = filtered_user_message_counts.median()

# –í—ã–≤–æ–¥–∏–º –æ–±–Ω–æ–≤–ª–µ–Ω–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏
print("–°—Ä–µ–¥–Ω–µ–µ —Å–æ–æ–±—â–µ–Ω–∏–π –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è (–±–µ–∑ –∏—Å–∫–ª—é—á–µ–Ω–Ω—ã—Ö):", mean_messages_per_user)
print("–ú–µ–¥–∏–∞–Ω–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–π –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è (–±–µ–∑ –∏—Å–∫–ª—é—á–µ–Ω–Ω—ã—Ö):", median_messages_per_user)

# –ë–µ—Ä—ë–º —Ç–æ–ø-10 –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
top_10_filtered_users = filtered_user_message_counts.head(10)

# 4. –¢–æ–ø-10 –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
plt.figure(figsize=(8, 6))
top_10_filtered_users.plot(kind="barh", color="#FF1493")
plt.axvline(mean_messages_per_user, color="gray", linestyle="--", label="–°—Ä–µ–¥–Ω–µ–µ")
plt.axvline(median_messages_per_user, color="blue", linestyle="-", label="–ú–µ–¥–∏–∞–Ω–∞")
plt.legend()
plt.title("–¢–æ–ø-10 –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (–±–µ–∑ —á–∞—Ç–æ–≤)", fontsize=14)
plt.xlabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–æ–±—â–µ–Ω–∏–π", fontsize=12)
plt.ylabel("–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å", fontsize=12)
plt.show()
```

# –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π —Å–µ–º–∞–Ω—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ "–æ—Ç—Ü–æ–≤—Å–∫–æ–≥–æ —è–∑—ã–∫–∞"

```python
import nltk
from nltk.corpus import stopwords

# 1. –ó–∞–≥—Ä—É–∂–∞–µ–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–µ —Ä—É—Å—Å–∫–∏–µ —Å—Ç–æ–ø-—Å–ª–æ–≤–∞ –∏–∑ NLTK
nltk.download("stopwords")
stop_words = set(stopwords.words("russian"))

# 2. –î–æ–±–∞–≤–ª—è–µ–º —Å–≤–æ–∏ —Å—Ç–æ–ø-—Å–ª–æ–≤–∞
custom_stop_words = {"—ç—Ç–æ", "–æ—Ç—Ü–æ–≤", "–∫–æ—Ç–æ—Ä—ã–µ", "–æ—á–µ–Ω—å", "—Ç–∞–∫–∂–µ", "–æ—Ç—Ü–∞", "–¥–µ—Ç–∏", "–æ—Ç–µ—Ü", "2", "1"}
stop_words.update(custom_stop_words)  # –î–æ–±–∞–≤–ª—è–µ–º –≤ –æ–±—â–∏–π —Å–ø–∏—Å–æ–∫

# 3. –û–±—ä–µ–¥–∏–Ω—è–µ–º –≤—Å–µ —Å–æ–æ–±—â–µ–Ω–∏—è –≤ –æ–¥–∏–Ω —Ç–µ–∫—Å—Ç
all_text = " ".join(df_otec["text"].dropna())

# 4. –û—á–∏—â–∞–µ–º —Ç–µ–∫—Å—Ç (—É–±–∏—Ä–∞–µ–º –∑–Ω–∞–∫–∏ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è, –ø–µ—Ä–µ–≤–æ–¥–∏–º –≤ –Ω–∏–∂–Ω–∏–π —Ä–µ–≥–∏—Å—Ç—Ä)
all_text = re.sub(r"[^\w\s]", "", all_text.lower())

# 5. –†–∞–∑–±–∏–≤–∞–µ–º —Ç–µ–∫—Å—Ç –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω—ã–µ —Å–ª–æ–≤–∞
words = all_text.split()

# 6. –§–∏–ª—å—Ç—Ä—É–µ–º —Å–ª–æ–≤–∞, —É–±–∏—Ä–∞—è **nltk-—Å—Ç–æ–ø-—Å–ª–æ–≤–∞ + —Å–≤–æ–∏**
filtered_words = [word for word in words if word not in stop_words]

# 7. –ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ–º —á–∞—Å—Ç–æ—Ç—É —Å–ª–æ–≤
word_counts = Counter(filtered_words)

# 8. –í—ã–±–∏—Ä–∞–µ–º —Ç–æ–ø-15 —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö —Å–ª–æ–≤
top_15_words = word_counts.most_common(15)

# 9. –í—ã–≤–æ–¥–∏–º –≤ –∫–æ–Ω—Å–æ–ª–∏
print("–¢–æ–ø-15 —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö —Å–ª–æ–≤")
for i, (word, freq) in enumerate(top_15_words, start=1):
    print(f"{i}. {word}: {freq}")
```

```python
# –°–æ—Ö—Ä–∞–Ω–∏–º –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
df_word_counts = pd.DataFrame(top_15_words, columns=["–°–ª–æ–≤–æ", "–ß–∞—Å—Ç–æ—Ç–∞"])
df_word_counts.to_csv("word_frequencies.csv", index=False, encoding="utf-8")
```
